## Helper Functions

```python
def get_missing_val_stats(df, need_plot = True, threshold = 0):
    '''
    This function calculates non-zero missing values and percentages for given dataframe and plot a seaborn bar plot on that.
    '''
    missing_count = df.isnull().sum()
        
    # We only interest in non-zero missing values here. Let's filter those ones
    non_zero_missing_count = missing_count[missing_count>0]
    non_zero_missing_pct = round(non_zero_missing_count/df.shape[0]*100,2)
    
    missing_stats_df = pd.DataFrame({'Null Count': non_zero_missing_count, 
                                     'Null Percentage': non_zero_missing_pct})
    
    print(f'Dataset has {df.shape[1]} features, out of which {df.shape[1]-non_zero_missing_count.shape[0]} have no missing values.')
    
    missing_stats_df.sort_values(by = 'Null Percentage', ascending = False, inplace = True)
    
    if need_plot:
        plt.figure(figsize = [10,5])
        sns.barplot(x = missing_stats_df.index, y = missing_stats_df['Null Percentage'])
        plt.xticks(rotation = 90)
        plt.show()
    
    # this is to show the columns those having missing value count > threshold. By default, we show any percentage > 0
    missing_stats_df = missing_stats_df[missing_stats_df['Null Percentage'] > threshold]
    return missing_stats_df
```

```python
def get_categorical_numerical_feature_split(df):
    '''
    This function prints the numerical and categorical features for a given dataframe.
    '''
    numerical_features = df.select_dtypes(exclude = ['object']).columns
    categorical_features = df.select_dtypes(include = ['object']).columns
    
    print('Categorical Features:\n', categorical_features)
    print('\n')
    print('Numerical Features:\n', numerical_features)    
```

```python
def get_highly_correlated_features(df, threshold = 0.8):
    """
    This function will take a dataframe and provides features that are correlated above the threshold
    """    
    feature_set = set()   # empty set, which will be populated in the loop to avoid double ups
    
    corr_df = df.corr() > threshold
    
    for i, feature in enumerate(corr_df.columns):
        # we get the indices where we have non-zero correlation, meaning above 0.8
        non_zero_elements = corr_df[feature].to_numpy().nonzero()

        # obviously, the feature with itself is having correlation as 1. So, we need to remove that index.
        non_zero_list = non_zero_elements[0].tolist()
        non_zero_list.remove(i)    

        if ((len(non_zero_list) != 0) & (feature not in feature_set)):
            feature_set.add(corr_df.columns[non_zero_list][0])
            print(feature, corr_df.columns[non_zero_list][0])
```

```python
def get_unique_values_for_features(df):
    unique_values = [] 
    
    for feature in df.columns:
        unique_values.append(df[feature].unique().shape[0])
    
    unique_values_df = pd.DataFrame({'Feature': df.columns, 'Unique Value Count': unique_values})
    unique_values_df.sort_values(by = 'Unique Value Count', ascending = True, inplace = True)
    return unique_values_df
```

## Data Clean-up

### Dropping Features with High Null Value Count

```python
get_missing_val_stats(train, need_plot=True, threshold=75)
```

```python
get_missing_val_stats(test, need_plot=True, threshold=75)
```

Looking at these stats, it appears that there are some extra columns in the test dataset, which has missing values compared to training dataset. However, there are 4 fields those having more than 75% of the data missing.

I will drop those features from our train and test datasets.

```python
train.drop(columns=['PoolQC', 'MiscFeature', 'Alley', 'Fence'], inplace = True)
test.drop(columns=['PoolQC', 'MiscFeature', 'Alley', 'Fence'], inplace = True)
```

### Imputing Missing Values

```python
# Let's now see how the missing value distribution again. 
missing_train = get_missing_val_stats(df = train, need_plot=False, threshold=0)
missing_train
```

```python
missing_test = get_missing_val_stats(df = test, need_plot=False, threshold=0)
missing_test
```

As a MVP model, let's look at the features missing in the test set and fill as below:
    * Categorical: with training data mode
    * Numerical: with training data mean

```python
feature_list = set(missing_test.index)
feature_list.update(set(missing_train.index))

categorical_features = []

for feature in feature_list:
    if (train[feature].dtype == 'object'):
        categorical_features.append(feature)
        
numerical_features = list(set(feature_list).difference(set(categorical_features)))
```

```python
# Replace each missing value for categorical features with mode.
for feature in categorical_features:
    train[feature].fillna(train[feature].mode()[0], inplace = True)
    test[feature].fillna(train[feature].mode()[0], inplace = True)
```

```python
# Replace each missing value for numerical features with mean.
for feature in numerical_features:
    train[feature].fillna(train[feature].mean(), inplace = True)
    test[feature].fillna(train[feature].mean(), inplace = True)
```

```python
get_missing_val_stats(df = train, need_plot=False, threshold=0)
```

```python
# Let's see how these features correlation with the target variable
from scipy.stats import pearsonr

# any value for p-value < 0.001 is significant predictor for target variable. 
feature_list = ['YearBuilt', 'GarageYrBlt', 'TotalBsmtSF', '1stFlrSF', 'GrLivArea', 'TotRmsAbvGrd', 'GarageCars', 'GarageArea']

for feature in feature_list:
    pair_df = train[[feature, 'SalePrice']].dropna()
    print(feature, pearsonr(pair_df[feature], pair_df['SalePrice']))
```

Looking at the correlation and p-value numbers here, all 8 features are statistically significant for the target variable (SalePrice). However, due to highly correlated nature of each pair of variables, it makes redundant to have all 8 features.

So, I will remove 4 features from the test and training datasets.

## Feature Selection for the MVP

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
X = train.iloc[:, :-1]
y = train.iloc[:, -1]
```

```python
# SelectKBest in sklearn to select 10 statistically important features (based on chi-squared test).
selector = SelectKBest(score_func=chi2, k=20)

# fit the dataset to the selector
X_new = selector.fit_transform(X,y)

# statistical test scores for features
scores = selector.scores_

# Let's create a dataframe for the results

df_scores = pd.DataFrame(data={'Feature': X.columns, 'Chi_2 Scores': scores})
df_scores.sort_values(by = 'Chi_2 Scores', ascending = False).reset_index(drop = True)

# As can be seen from below dataframe, there are certain features with very high statistical score. Those high valued
# features are selected by the algorithm.

# X_new contains the extracted dataset. 
```
